```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# Generate 2D dataset
def generate_2d_data(n_samples, noise):
    X1 = np.random.randn(n_samples, 2) + np.array([0, 0])
    X2 = np.random.randn(n_samples, 2) + np.array([4, 4])
    X = np.vstack((X1, X2))
    y = np.hstack((np.zeros(n_samples), np.ones(n_samples)))
    
    noise = np.random.normal(0, noise, X.shape)
    X = X + noise
    
    return X, y

# Generate 3D dataset
def generate_3d_data(n_samples, noise):
    X1 = np.random.randn(n_samples, 3) + np.array([0, 0, 0])
    X2 = np.random.randn(n_samples, 3) + np.array([4, 4, 4])
    X = np.vstack((X1, X2))
    y = np.hstack((np.zeros(n_samples), np.ones(n_samples)))
    
    noise = np.random.normal(0, noise, X.shape)
    X = X + noise
    
    return X, y

n_samples = 100
noise = 1.2  # Increased noise

# 2D dataset
X_2d, y_2d = generate_2d_data(n_samples, noise)
y_2d[y_2d == 0] = -1
data_2d = np.column_stack((X_2d, y_2d))
df_2d = pd.DataFrame(data_2d, columns=['X1', 'X2', 'y'])
df_2d.to_csv('2d_noise_data.csv', index=False)

# 3D dataset
X_3d, y_3d = generate_3d_data(n_samples, noise)
y_3d[y_3d == 0] = -1
data_3d = np.column_stack((X_3d, y_3d))
df_3d = pd.DataFrame(data_3d, columns=['X1', 'X2', 'X3', 'y'])
df_3d.to_csv('3d_noise_data.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

n_samples = 100
n_features = 1
noise = 10

X_simple, y_simple = make_regression(n_samples=n_samples, n_features=n_features, noise=noise, random_state=42)

# Adding Non-linearity
y_simple = y_simple + np.sin(X_simple).ravel() * 40

# plot the data
plt.scatter(X_simple, y_simple)
plt.show()

# Save the data
import pandas as pd

data_simple = np.column_stack((X_simple, y_simple))
df_simple = pd.DataFrame(data_simple, columns=['X', 'y'])
df_simple.to_csv('simple_data.csv', index=False)

```

```python
n_samples = 100
n_features = 2
noise = 10

X_multiple, y_multiple = make_regression(n_samples=n_samples, n_features=n_features, noise=noise, random_state=42)

# Adding Non-linearity
y_multiple = y_multiple + np.sin(X_multiple[:,0]).ravel() * 40 + np.cos(X_multiple[:,1]).ravel() * 40

# plot the data
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_multiple[:,0], X_multiple[:,1], y_multiple)
plt.show()

# Save the data
import pandas as pd

data_multiple = np.column_stack((X_multiple, y_multiple))
df_multiple = pd.DataFrame(data_multiple, columns=['X1', 'X2', 'y'])
df_multiple.to_csv('multiple_data.csv', index=False)
```
